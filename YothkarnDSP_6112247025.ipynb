```python
#อธิบายคำสั่งการใช้งาน
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display

```

```python
#lab01
sin_sig = thinkdsp.SinSignal(freq=7025, amp=25,offset=0)
sin_sig.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
#lab02
signal = thinkdsp.TriangleSignal(2500)
duration = signal.period*25
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
snwave = signal.make_wave(duration=25, framerate=10000)
snwave.apodize()
snwave.make_audio()
```

```python
spectrum = snwave.make_spectrum()
spectrum.plot()
```

```python
signal = thinkdsp.SquareSignal(25)
duration = signal.period*25
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)
sqwave = signal.make_wave(duration=25, framerate=10000)
sqwave.apodize()
sqwave.make_audio()
```
